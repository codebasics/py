```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('Churn_Modelling.csv')
df
```

```python
df1 = df.drop(['CustomerId', 'Surname', 'RowNumber'], axis='columns')
df1
```

```python
Fields_toScale = ['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'EstimatedSalary']
Fields_toencode = ['Geography']
object_tobinary = ['Gender']
```

```python
df1['Gender'] = df1.Gender.replace({'Female': 1, 'Male' : 0})
```

```python
scaler = MinMaxScaler()
df1[Fields_toScale] = scaler.fit_transform(df1[Fields_toScale])
```

```python
df1 = pd.get_dummies(data=df1, columns=Fields_toencode)
df1
```

```python
X = df1.drop('Exited', axis='columns')
Y = df1.Exited

x_train, x_test, y_train, y_test = train_test_split(X, Y, train_size=0.9, random_state=10)
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
x_train.shape
```

```python
model = keras.Sequential([
    keras.layers.Dense(20, input_shape=(12,), activation='relu'),
    keras.layers.Dense(10, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer = 'adam',
              loss = 'binary_crossentropy',
              metrics = ['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=100)
```

```python
model.evaluate(x_test, y_test)
```

```python
yp = model.predict(x_test)
```

```python
y_pred = []
for i in yp:
    if i > 0.5:
        y_pred.append(1)
    else:
        y_pred.append(0)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print(classification_report(y_test, y_pred))
```

## Sampling Data:

```python
def ANN(x_train, x_test, y_train, y_test, epochs, loss_fn):
    model = keras.Sequential([
        keras.layers.Dense(20, input_shape=(12,), activation='relu'),
        keras.layers.Dense(10, activation='relu'),
        keras.layers.Dense(1, activation='sigmoid')
    ])

    model.compile(optimizer = 'adam',
              loss = loss_fn,
              metrics = ['accuracy'])
    
    model.fit(x_train, y_train, epochs=epochs)
    
    print(model.evaluate(x_test, y_test))
    
    yp = model.predict(x_test)
    y_pred = np.round(yp)
    
    print(classification_report(y_test, y_pred))
    return y_pred
```

## Under Sampling:

```python
df_count_1 = df1[df1['Exited'] == 1]
df_count_0 = df1[df1['Exited'] == 0].sample(df_count_1.shape[0], replace=True)
df_count_0.shape
```

```python
df_under = pd.concat([df_count_0, df_count_1], axis=0)
df_under.shape
```

```python
X = df_under.drop('Exited', axis='columns')
Y = df_under.Exited

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size=0.9, random_state=10, stratify=Y)
```

```python
ANN(X_train, X_test, Y_train, Y_test, 100, 'binary_crossentropy')
```

## OverSampling:

```python
df_count_1 = df1[df1['Exited'] == 1].sample(df_count_0.shape[0], replace=True)
df_count_0 = df1[df1['Exited'] == 0]
```

```python
df_over = pd.concat([df_count_0, df_count_1], axis=0)
df_over.shape
```

```python
ANN(X_train, X_test, Y_train, Y_test, 100, 'binary_crossentropy')
```
