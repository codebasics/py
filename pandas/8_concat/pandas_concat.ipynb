## <font color='blue'>Basic Concatenation</font>

```python
import pandas as pd

india_weather = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
    "humidity": [80, 60, 78]
})
india_weather
```

```python
us_weather = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
    "humidity": [68, 65, 75]
})
us_weather
```

```python
#This adds the us weather df after india weather df
df = pd.concat([india_weather, us_weather])
df
#But the index here is messy
```

## <font color='blue'>Ignore Index</font>

```python
df = pd.concat([india_weather, us_weather], ignore_index=True)
df
```

## <font color='blue'>Concatenation And Keys</font>

```python
#Associating a key with a dataframe
df = pd.concat([india_weather, us_weather], keys=["india", "us"])
df
```

```python
df.loc["us"]
```

```python
df.loc["india"]
```

## <font color='blue'>Concatenation Using Index</font>

```python
temperature_df = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
}, index=[0,1,2]) #creating these index will help further concatenating
temperature_df
```

```python
windspeed_df = pd.DataFrame({
    "city": ["delhi","mumbai"],
    "windspeed": [7,12],
}, index=[1,0]) #forcing delhi to have index of 1
windspeed_df
```

```python
#You want windspeed to appear as an individual column
#axis=0, means appendig the second dataframe as rows; but axis=1 means appending as column
df = pd.concat([temperature_df,windspeed_df],axis=1)
df
```

## <font color='blue'>Concatenate dataframe with series</font>

```python
s = pd.Series(["Humid","Dry","Rain"], name="event")
s
```

```python
df = pd.concat([temperature_df,s],axis=1)
df
```
