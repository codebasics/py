## **Dataframe is most commonly used object in pandas. It is a table like datastructure containing rows and columns similar to excel spreadsheet**

```python
import pandas as pd
weather_data = {
    'day': ['1/1/2017','1/2/2017','1/3/2017','1/4/2017','1/5/2017','1/6/2017'],
    'temperature': [32,35,28,24,32,31],
    'windspeed': [6,7,2,7,4,2],
    'event': ['Rain', 'Sunny', 'Snow','Snow','Rain', 'Sunny']
}
df = pd.DataFrame(weather_data)
df = pd.read_csv("weather_data.csv")
df
```

## <font color='blue'>Operations On DataFrame</font>

```python
df
```

```python
#We are extracing a single maximum value from the temperature column in the dataframe
df['temperature'].max()
```

```python
# Extracing the minimum windspeed
df['windspeed'].min()
```

```python
#This is extracing a row whose temperature is greater than 32
df[df['temperature']>32]
```

```python
# Extracing a row whose event value is Rain
df[df['event']=='Rain']
```

```python
# This is extracing a row whose temperature has the maximum value
# This is different from just extracing a single value
df[df['temperature'] == df['temperature'].max()] # Kinda doing SQL in pandas
```

```python
#You could also print a certain column from this specificaiton
df['day'][df['temperature'] == df['temperature'].max()] # Kinda doing SQL in pandas
```

```python
# P: select the windspeed when the event is Rain
df['windspeed'][df['event']=='Rain']
```

```python
# P: select the windspeed when the temperature is at minimum value
df['windspeed'][df['temperature']==df['temperature'].min()]
```

```python
# Selecting a row whose temperature is at maximum value
df[df['temperature'] == df['temperature'].max()] # Kinda doing SQL in pandas
```

```python
df['temperature'].std()
```

```python
df['event'].max() # But mean() won't work since data type is string
```
