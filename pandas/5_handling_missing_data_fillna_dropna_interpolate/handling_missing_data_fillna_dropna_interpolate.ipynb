## <font color="maroon"><h4 align="center">Handling Missing Data - fillna, interpolate, dropna</font>

```python
import pandas as pd
df = pd.read_csv("weather_data.csv",parse_dates=['day'])
type(df.day[0])
df
```

```python
df.set_index('day',inplace=True)
df
```

## <font color="blue">fillna</font>

<font color="purple">**Fill all NaN with one specific value**</font>

```python
new_df = df.fillna(0)
new_df
```

<font color="purple">**Fill na using column names and dict**</font>
- This allows you to avoid filling in certain columns

```python
new_df = df.fillna({
        'temperature': 0,
        'windspeed': 0,
        'event': 'No Event'
    })
new_df
```

<font color="purple">**Use method to determine how to fill na values**</font>

```python
#forward fill means if there's any missing value, then just carry forward the previous value in the df
new_df = df.fillna(method="ffill")
new_df
```

```python
#backward fill
new_df = df.fillna(method="bfill")
new_df
```

<font color="purple">**Use of axis**</font>
- With axis, you can specify which direction to fill the missing values

```python
#for example this is backward filling data in the same row from the next column
new_df = df.fillna(method="bfill", axis="columns") # axis is either "index" or "columns"
new_df
```

<font color="purple">**limit parameter**</font>
- We can specify we just want to carry forward the value by only one row if there's more missing values

```python
new_df = df.fillna(method="ffill",limit=1)
new_df
```

### <font color="blue">interpolate</font>
- Basically it is linear interpolation

```python
new_df = df.interpolate()
new_df
```

```python
#you can use time interpolate: it makes more sense as it considers time
new_df = df.interpolate(method="time") 
new_df
```

**Notice that in above temperature on 2017-01-04 is 29 instead of 30 (in plain linear interpolate)**

**There are many other methods for interpolation such as quadratic, piecewise_polynomial, cubic etc. 
Just google "dataframe interpolate" to see complete documentation**

### <font color="blue">dropna</font>

```python
# It drops rows with at least one na value
new_df = df.dropna()
new_df
```

```python
#It drops values with all values with na
new_df = df.dropna(how='all')
new_df
```

```python
#If we have at least one "not na value", then we keep that row
new_df = df.dropna(thresh=1)
new_df
```
